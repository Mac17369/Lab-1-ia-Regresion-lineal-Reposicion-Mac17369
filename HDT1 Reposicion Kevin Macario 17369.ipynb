### Se genera el modelo a partir de un Training Set de datos y, se realiza con modelos demostrados en clase: funci√≥n costo, su derivada y el descenso al gradiente. Con esto se obtienen las thetas y se determina el respectivo Lambda

```python
TRAINING_ELEMENTS = 240
TEST_ELEMENTS = 80

datosUtilizar = pd.read_csv('Admission_Predict.csv', usecols=['GRE Score', 'TOEFL Score', 'Chance of Admit '])
datosValues = datosUtilizar.iloc[:len(datosUtilizar), 0:3].values
datosValues

training = datosValues[:TRAINING_ELEMENTS]
test = datosValues[TRAINING_ELEMENTS+TEST_ELEMENTS:]

trainingPre = training[:len(datosUtilizar), 2].reshape(-1,1)
testPre = test[:len(datosUtilizar), 2].reshape(-1,1)

X = (datosUtilizar.iloc[:len(datosUtilizar), 0].values )
X = X.reshape(-1,1)
X = np.insert(X, 0, (datosUtilizar.iloc[:len(datosUtilizar), 0].values ) , axis = 1)
X = np.insert(X, 0, 1, axis = 1)
y = datosUtilizar.iloc[:len(datosUtilizar), 2].values.reshape(-1,1)

m, n = X.shape

lambda_x = 0

theta_0 = np.random.rand(n, 1)
theta, costs, gradient_norms = gradient_descent(
    X,
    y,
    theta_0,
    linear_cost,
    linear_cost_derivate,
    lambda_x,
    alpha=0.000001,
    treshold=0.0001,
    max_iter=10000
)

train = []
cv = []

for i in range(0, 100, 10):
    train.append(linear_cost(training, trainingPre, theta, i))
    cv.append(linear_cost(test, testPre, theta, i))

plt.plot(np.arange(0, 100, 10), train, color="yellow")
plt.plot(np.arange(0, 100, 10), cv, color="blue")
```

<div style="text-align: justify ">
CONCLUSIONES:
Cuando se incrementa el valor de labda los valores de costo no convergen. Esto representa un problema BIAS. Se debe de mejorar el Lambda
</div>

### CPGA vs Chance of Admit

```python
TRAINING_ELEMENTS = 240

x1 = np.asarray(cgpa[:TRAINING_ELEMENTS])

X = np.vstack(
    (
        np.ones(TRAINING_ELEMENTS),
        x1,
        (x1 ** 4)/1000
    )
).T


y = np.asarray(chance_of_admit[:TRAINING_ELEMENTS])

dataset_1 = (X, y.reshape(TRAINING_ELEMENTS, 1))
```

#### Grafica de costos

```python
(X, y) = dataset_1
m, n = X.shape

theta_0 = np.random.rand(n, 1)
lambda_x = 10
theta, costs, gradient_norms = gradient_descent(
    X,
    y,
    theta_0,
    linear_cost,
    linear_cost_derivate,
    lambda_x,
    alpha=0.00001,
    treshold=0.0001,
    max_iter=300000
)

print ('THETA:', theta)

plt.plot(np.arange(len(costs)), costs)
plt.show()

plt.scatter(X[:, 1], y)
plt.scatter(X[:, 1], np.matmul(X, theta), color='red')
plt.show()

```

CONCLUSIONES:
Al buscar el menor costo se obtuvo valores de theta que brindan una funcion que se acopla a los que se esperaba obtener con las predicciones para la varuable CPGA

### CGPA, GRE Score, TOEFL Score vs Chance of Admit

```python
TRAINING_ELEMENTS = 240
X = np.vstack(
    (
        np.ones(TRAINING_ELEMENTS),
        (x1 ** 3)/1000,
        (x2 ** 2)/100,
        (x3 ** 2)/100
    )
).T

y = np.asarray(chance_of_admit[:TRAINING_ELEMENTS])

dataset_4 = (X, y.reshape(TRAINING_ELEMENTS, 1))
```

```python
(X, y) = dataset_4
m, n = X.shape

theta_0 = np.random.rand(n, 1)
lambda_x = 10
theta, costs, gradient_norms = gradient_descent(
    X,
    y,
    theta_0,
    linear_cost,
    linear_cost_derivate,
    lambda_x,
    alpha=0.000001,
    treshold=0.0001,
    max_iter=300000
)

print ('THETA:', theta)

plt.plot(np.arange(len(costs)), costs)
plt.show()

plt.scatter(X[:, 1], y)
plt.scatter(X[:, 1], np.matmul(X, theta), color='red')
plt.show()
```

## Se comprueba la regresi√≥n lineal obtenida

```python
print ('THETA:', theta)

TRAINING_ELEMENTS = 240
TEST_ELEMENTS = 80

t1 = np.asarray(cgpa[TRAINING_ELEMENTS:TRAINING_ELEMENTS + TEST_ELEMENTS])
t2 = np.asarray(gre_score[TRAINING_ELEMENTS:TRAINING_ELEMENTS + TEST_ELEMENTS])
t3 = np.asarray(toefl_score[TRAINING_ELEMENTS:TRAINING_ELEMENTS + TEST_ELEMENTS])

X = np.vstack(
    (
        np.ones(TEST_ELEMENTS),
        (t1 ** 3)/1000,
        (t2 ** 2)/100,
        (t3 ** 2)/100
    )
).T

y = np.asarray(chance_of_admit[TRAINING_ELEMENTS:TRAINING_ELEMENTS + TEST_ELEMENTS])
y = y.reshape(TEST_ELEMENTS, 1)

plt.scatter(X[:, 1], y, color='blue')
plt.scatter(X[:, 1], np.matmul(X, theta), color='red')
plt.show()
```

CONCLUSIONES: A pesar que algunos puntos no han sido posibles de predecir, este algoritmo ya que estos puntos han sido anormales y no se utilizaron para el entrenamiento e incluso se acercan. Ademas, el algoritmo no dio puntos anormales y las predicciones fueron exactas y precisas y no se salio del rango establecido 0 < x < 1. El modelo se acepta y se procede con CROSS VALIDATION TEST

## Cross Validation Test 

```python
print ('THETA:', theta)
TRAINING_ELEMENTS = 240
TEST_ELEMENTS = 80
CROSS_VAL_ELEMENTS = 80

t1 = np.asarray(cgpa[TRAINING_ELEMENTS + TEST_ELEMENTS:TRAINING_ELEMENTS + TEST_ELEMENTS + CROSS_VAL_ELEMENTS])
t2 = np.asarray(gre_score[TRAINING_ELEMENTS + TEST_ELEMENTS:TRAINING_ELEMENTS + TEST_ELEMENTS + CROSS_VAL_ELEMENTS])
t3 = np.asarray(toefl_score[TRAINING_ELEMENTS + TEST_ELEMENTS:TRAINING_ELEMENTS + TEST_ELEMENTS + CROSS_VAL_ELEMENTS])

X = np.vstack(
    (
        np.ones(CROSS_VAL_ELEMENTS),
        (t1 ** 3)/1000,
        (t2 ** 2)/100,
        (t3 ** 2)/100
    )
).T

y = np.asarray(chance_of_admit[TRAINING_ELEMENTS+ TEST_ELEMENTS:TRAINING_ELEMENTS + TEST_ELEMENTS + CROSS_VAL_ELEMENTS])
y = y.reshape(CROSS_VAL_ELEMENTS, 1)

plt.scatter(X[:, 1], y, color='blue')
plt.scatter(X[:, 1], np.matmul(X, theta), color='red')
plt.show()
```

CONLCUSIONES: se reafirma que el modelo usado fue bueno, ya que logro predecir muy bien los puntos, las predicciones no tan aproximadas no se exedieron de manera significativa el rango de prediccion.
